## Program 1: Swap Two Numbers

**Problem**: Read two integers `a` and `b` and swap them.

**Input**:
- Two integers `a` and `b` (space-separated)

**Output**:
- Print the swapped values as: `a b`

**Sample Input**:
```

```
**Sample Output**:



````python
a, b = map(int, input().split())
print(f"{a} {b}")
a, b = b, a
print(f"{a} {b}")
````

## Program 4: Simple Interest

**Problem**: Read `P`, `R`, `T` and compute Simple Interest.

**Formula**: SI = (P * R * T) / 100

- Three numbers `P R T` (space-separated)

**Output**:
- Print `SI`

```
1000 5 2
```


````python
principal= float(input("Enter principal amount: "))
rate= float(input("Enter rate of interest: "))
time= float(input("Enter time in years: "))
print(f"Principal={principal:.2f} Rate={rate:.2f} Time={time:.2f}")
simple_interest = (principal * rate * time)/100
print(f"Simple interest = {simple_interest:.1f}")
````

## Program 8: Leap Year Check

**Problem**: Read a year and print `Leap Year` or `Not Leap Year`.

**Rule**:
- Leap if divisible by 400 OR divisible by 4 but not by 100.

**Input**:
- One integer `year`

**Output**:
- `Leap Year` or `Not Leap Year`

**Sample Input**:
```
2024
```
**Sample Output**:
```
Leap Year
```


````python
try:
    input_year = int(input("Enter a year:"))
    if((input_year % 4 ==0 and input_year %100 !=0) or (input_year % 400 ==0)):
        print(f"The year {input_year} is a Leap Year")
    else:
        print(f"The year {input_year} is not a Leap Year")
except ValueError:
    print("Invalid input! Please enter a valid year.")
````

## Program 9: Grade from Marks

**Problem**: Read marks (0–100) and print grade.

**Grading**:
- 90–100: A
- 80–89: B
- 70–79: C
- 60–69: D
- <60: F

**Input**:
- One integer `marks`

**Output**:
- Print the grade letter

**Sample Input**:
```
83
```
**Sample Output**:
```
B
```


````python
try:
    marks = int(input("Enter marks between 0 and 100:"))
    if 90<=marks<=100:
        grade = 'A'
    elif 80<=marks<90:
        grade = 'B'
    elif 70<=marks<80:
        grade = 'C'
    elif 60<=marks<70:
        grade = 'D'
    elif 0<=marks<60:
        grade = 'F'
    else:
        grade = None
        print("Marks should be between 0 and 100.")
    if grade:
        print(f"Grade: {grade}")
except ValueError:
    print("Invalid input! Please enter valid marks.")
````

## Program 11: Sum of First N Natural Numbers

**Problem**: Read `N` and print the sum 1 + 2 + ... + N.

- One integer `N`


````python
try:
    input_integer = int(input("Enter an integer:"))
    print(f"Input integer: {input_integer}")
    result =0
    for i in range(1,input_integer + 1):
        result += i
    print(f"The sum of first {input_integer} natural numbers is {result}")
except ValueError:
    print("Invalid input! Please enter a valid integer.")
````
